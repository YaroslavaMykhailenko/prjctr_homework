    You should show user a random country from the list and ask him to guess the capital. If user input right capital, print "You are right!", add him a point and ask him to guess another country. If not, you should ask again. User should be able to quit the game by typing "exit".
    You should print current score after each round. Also, you should print the final score before user quit the game.

````python
# + If user make a mistake you should decrement his lives. 
# Initial amount of lives is 3. Game will end when user has no lives left.
# You should print the final score after user has no lives left.
import random
capitals = {
        'Ukraine': 'Kyiv',
        'France': 'Paris',
        'Germany': 'Berlin',
        'Italy': 'Rome', 
        'USA': 'Washington',
        'Canada': 'Ottawa',
        'Switzerland': 'Bern',
        'Austria': 'Vienna',
        'Belgium': 'Brussels',
        'Sweden': 'Stockholm',
        'Norway': 'Oslo',
        'Denmark': 'Copenhagen',
        'Finland': 'Helsinki',
        'Poland': 'Warsaw',
        'Romania': 'Bucharest',
        'Bulgaria': 'Sofia',
        'Greece': 'Athens'
    }



def guess_capital(dictionary: dict) -> int:

    score = 0
    life_score = 3
      
    rand_country = random.choice(list(dictionary.keys()) )
    while True:       
        print('Country:', rand_country)         

        user_input = input(f'What is the capital of {rand_country}?:').lower().capitalize()

        if user_input == 'Exit':
            break

        if dictionary[rand_country] != user_input:            
            print('Your answer is wrong, try to guess one more time!')
            life_score -= 1
            if life_score == 0:
                print(f'You lose!')    
                break     
            
        else:
            score += 1
            print(f'You are right. Now your score is {score}')
            rand_country = random.choice(list(dictionary.keys()) )        

            

    print(f'Thank you for the game. Your general score is {score}')

    return score    

    

guess_capital(capitals)
````
