Create table
```sql

CREATE TABLE users(
    id serial PRIMARY KEY,
    name VARCHAR(50) NOT NULL,
    gender VARCHAR(50) NOT NULL,
    phone_number VARCHAR(50) NOT NULL,
    email_adress VARCHAR(50) NOT NULL
);

CREATE TABLE hosts(
    id serial PRIMARY KEY,
	user_id INTEGER REFERENCES users(id)      
);

CREATE TABLE guests(
    id serial PRIMARY KEY,
	user_id INTEGER REFERENCES users(id)
);

CREATE TABLE rooms(
    id serial PRIMARY KEY,
	host_id INTEGER REFERENCES hosts(id),
    amount_of_residents INT NOT NULL,
    price INT NOT NULL,
    fridge BOOLEAN NOT NULL,
    quantity_of_beds INT NOT NULL,
    current_adress VARCHAR(50) NOT NULL
);

CREATE TABLE reservation(
    id serial PRIMARY KEY,
    guest_id INTEGER REFERENCES guests(id),
	room_id INTEGER REFERENCES rooms(id),
    confirmation_status BOOLEAN NOT NULL,
    duration VARCHAR(50) NOT NULL,
    editional_information VARCHAR(50),
    total_price INT NOT NULL
);

CREATE TABLE booking(
    id serial PRIMARY KEY,
	reservation_id INTEGER REFERENCES reservation(id),
    booking_from VARCHAR(50) NOT NULL,
    booking_to VARCHAR(50) NOT NULL	
);

CREATE TABLE reviews(
    id serial PRIMARY KEY,
	guest_id INTEGER REFERENCES guests(id),
	host_id INTEGER REFERENCES hosts(id),
    rating INT NOT NULL,
    review_message VARCHAR(50)
);

CREATE TABLE payments(
    id serial PRIMARY KEY,
    reservation_id INTEGER REFERENCES reservation(id),
    payment_status BOOLEAN NOT NULL,
    payment_method VARCHAR(50) NOT NULL
	
);

```


```sql
-- 1. Find a user who had biggest amount of reservation. Return user name and user_id
SELECT guests.id as user_id, users.name as user_name, max(reservation.total_price) as biggest_reservation_amount FROM users 
INNER JOIN guests ON users.id = guests.user_id 
INNER JOIN reservation ON guests.id = reservation.guest_id 
GROUP BY users.name, guests.id, reservation.total_price
ORDER BY total_price DESC  LIMIT(1);

-- (Optional) Find a host who earned biggest amount of money for the last month. Return host name and host_id
SELECT hosts.id as host_id, users.name as user_name, reservation.room_id, reservation.total_price, sum(total_price) as earned_amount FROM reservation 
INNER JOIN rooms ON reservation.room_id = rooms.id 
INNER JOIN hosts ON rooms.host_id = hosts.id 
INNER JOIN users ON users.id = hosts.user_id 
GROUP BY users.name, hosts.id, reservation.room_id, reservation.total_price 
ORDER BY reservation.total_price 
DESC LIMIT(1);

--(Optional) Find a host with best average rating. Return host name and host_id
SELECT users.name as host_name, hosts.id as host_id, avg(reviews.rating) FROM users 
INNER JOIN hosts ON hosts.user_id = users.id 
INNER JOIN reviews ON hosts.id = reviews.host_id 
GROUP BY users.name, hosts.id 
ORDER BY avg(reviews.rating) 
DESC LIMIT(1);

```
